```python
import numpy as np 
import pandas as pd
import matplotlib.pylab as plt
%matplotlib inline

local_path = './kscoreVSothers.csv'

dfTW= pd.read_csv(local_path)

dfTW.head()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 8))
axes[0].scatter(dfTW["kscore"], dfTW["totalERA"], color = "steelblue")
axes[0].grid(alpha=0.25)
axes[0].set_xlabel("K Score", fontsize=16)
axes[0].set_ylabel("ERA", fontsize=16)
axes[0].set_title("K Score vs ERA", fontsize=16)

axes[1].scatter(dfTW["kscore"], dfTW["totalhits"], color = "steelblue")
axes[1].set_title("K Score vs Total Hits Allowed", fontsize=18)
axes[1].grid(alpha=0.25)
axes[1].set_xlabel("K Score", fontsize=16)
axes[1].set_ylabel("Hits allowed", fontsize=16)
```

```python
dfTW['kscore'].corr(dfTW['totalERA'], method='pearson')
```
